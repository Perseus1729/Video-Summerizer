```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url):
  return url.split('watch?v=')[-1]
```

```python
video_id_= get_video_id("https://www.youtube.com/watch?v=xDQL3vWwcp0&list=PL49M3zg4eCviRD4-hTjS5aUZs3PzAFYkJ")
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id_)
```

```python
transcript
```

```python
transcript_join = "".join([i['text'] for i in transcript])
```

```python
from rpunct import RestorePuncts
# The default language is 'english'
rpunct = RestorePuncts()
```

```python
result = rpunct.punctuate(transcript_join)
# print(result)
```

```python
import openai
```

```python
api_key = ""
```

```python
prompt = f"Summerize this transcript :- {result} "
```

```python
api_key1 = "" //From Open Router AI 
base_url = "https://openrouter.ai/api/v1"
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key1,base_url=base_url)
full_chat_ = client.chat.completions.create(
  messages = [
      {
          "role": "user",
          "content": prompt
      }

  ],
  model = "mistralai/mistral-7b-instruct",
  temperature = 1,
  max_tokens = 256,
  top_p = 1,
  frequency_penalty = 0,
  presence_penalty = 0
)
```
